```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

titanic = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataMining/framingham.csv");
print(titanic.shape[0])
print(list(titanic.columns))
titanic.isnull().values.any()
titanic.isnull().sum()
df_no_na = titanic.dropna()
print(df_no_na.shape[0])
((titanic["age"] < 0) | (titanic["age"] > 100)).values.any()
((titanic["totChol"] < 0) | (titanic["totChol"].isnull())).values.any()
titanic["totChol"].mean()
#titanic[["Age", "Fare"]].mean()

#10
plt.hist(titanic, bins=10, edgecolor="black")
plt.title("Simulated data")
plt.xlabel("x")
plt.ylabel("totChol")
plt.grid(axis="y", alpha=0.75)
plt.grid(axis="x", alpha=0.75)
plt.show()


plt.clf() # clean window
plt.boxplot(titanic,vert=0)
plt.title("Simulated data")
ax = plt.gca()
ax.set_yticklabels([])
plt.show()

#11
titanic.groupby("male")["male"].count()

#12 Calculate the mean of the total cholesterol, per gender
titanic.groupby("male")["totChol"].mean()

#13. Generate the boxplots for the cigarettes per day for males and females. Compare the
#distributions. Hint: you can use the boxplot from seaborn python package.
# Generate boxplots
sns.boxplot(x='male', y='cigsPerDay', data=titanic)

# Set plot labels and title
plt.xlabel('Sex')
plt.ylabel('Cigarettes per Day')
plt.title('Boxplots of Cigarettes per Day by Sex')

# Show the plot
plt.show()

#14. How can you visually study the relationship between total cholesterol and systolic blood
#pressure?
# Create a scatter plot
plt.scatter(titanic['totChol'], titanic['sysBP'])

# Set plot labels and title
plt.xlabel('Total Cholesterol')
plt.ylabel('Systolic Blood Pressure')
plt.title('Scatter Plot of Total Cholesterol vs. Systolic Blood Pressure')

# Show the plot
plt.show()


```

```python
def func1(n):
    sequence = [0]
    num = 1
    while num < n:
        sequence.append(num)
        num *= 2
    return sequence

number = int(input("please input integer："))
result = func1(number)
print(result)
print(','.join(map(str, result)))
```

```python
def func2(num, size):
    numbers = list(range(size))
    print("Generated list of numbers:", numbers)

    if num in numbers:
        return 'Found'
    else:
        return 'NotFound'

number = int(input("Enter a number to search: "))
result = func2(number, 10)
print(result)
```
